```python
import tensorflow as tf
import numpy as np
# define shape
m = 2  # number of examples (rows)
n = 3  # input features (columns)

# create matrix U
U = tf.constant([[1, 2, 3], [4, 5, 6]], dtype=tf.float32)
print("Matrix U:")
print(U)
```

```python
X = tf.transpose(U)
print("Matrix X (Transpose of U):")
print(X)
```

```python
Y = tf.random.uniform(shape=(1, m), minval=0, maxval=10, dtype=tf.int32)
print("Matrix Y:")
print(Y.numpy())
```

```python
p = 3  # number of hidden layer units
W1 = tf.Variable(tf.random.uniform(shape=(p, n), minval=0, maxval=10))
print("Matrix W1:")
print(W1.numpy())
```

```python
B1 = tf.Variable(tf.random.uniform((p, 1), 0, 1))#shape minval maxval
print("Vector B1:")
print(B1.numpy())
```

```python
W2 = tf.Variable(tf.zeros((1, p)), dtype=tf.float32)
print("Vector W2:")
print(W2.numpy())
```

```python
B2 = tf.Variable(tf.random.uniform((1, 1), 0, 1))
print("Scalar B2:")
print(B2.numpy())
```

```python
def relu(z):
    return tf.maximum(0.0, z)

def relu_prime(z):
    return tf.cast(z > 0, tf.float32)

def sigmoid(z):
    return 1 / (1 + tf.exp(-z))

def sigmoid_prime(z):
    g = sigmoid(z)
    return g * (1 - g)
```

```python
alpha = tf.constant(0.01)  # learning rate
num_iters = 15             # number of iterations

for i in range(num_iters):
    # Forward pass
    Z1 = tf.matmul(W1, X) + B1
    A1 = relu(Z1)
    Z2 = tf.matmul(W2, A1) + B2
    A2 = sigmoid(Z2)

    # Compute loss (L)
    L = 0.5 * tf.reduce_mean((A2 - tf.cast(Y, tf.float32)) ** 2)

        # Backward pass (gradients)
        # Backward pass (gradients)
    dA2 = A2 - tf.cast(Y, tf.float32)
    dZ2 = dA2 * sigmoid_prime(Z2)
    dA1 = tf.matmul(tf.transpose(W2), dZ2)
    dZ1 = dA1 * relu_prime(Z1)
    
    dW1 = (1 / m) * tf.matmul(dZ1, tf.transpose(X))
    dB1 = (1 / m) * tf.reduce_sum(dZ1, axis=1, keepdims=True)
    
    dW2 = (1 / m) * tf.matmul(dZ2, tf.transpose(A1))
    dB2 = (1 / m) * tf.reduce_sum(dZ2, axis=1, keepdims=True)  # âœ… FIXED

    # Parameter update
    W1.assign_sub(alpha * dW1)
    B1.assign_sub(alpha * dB1)
    W2.assign_sub(alpha * dW2)
    B2.assign_sub(alpha * dB2)

    # Parameter update
    W1.assign_sub(alpha * dW1)
    B1.assign_sub(alpha * dB1)
    W2.assign_sub(alpha * dW2)
    B2.assign_sub(alpha * dB2)

    # Display
    print(f"\nIteration {i + 1}")
    print("Loss:", L.numpy())
    print("W1:\n", W1.numpy())
    print("B1:\n", B1.numpy())
    print("W2:\n", W2.numpy())
    print("B2:\n", B2.numpy())
    
print("B2 shape:", B2.shape)
print("dB2 shape:", dB2.shape)
```
